```python
from skimage import filters, exposure
from matplotlib import pyplot as plt

image_col = cv2.imread('/Users/frodo/Desktop/1529528700_0_2018-06-20-22-05-00.jpg')
image = cv2.cvtColor(image_col, cv2.COLOR_BGR2GRAY)
threshold = filters.threshold_otsu(image)
mask = image < threshold
hist, bins_center = exposure.histogram(image)
plt.imshow(mask, cmap='gray', interpolation='nearest')
```

```python
ret,thresh = cv2.threshold(image,127,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
(img, cnts, _) = cv2.findContours(thresh.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for c in cnts:    
    cv2.drawContours(image_col, c, -1, (255, 0, 0), 3)
io.imshow(image_col)
```
